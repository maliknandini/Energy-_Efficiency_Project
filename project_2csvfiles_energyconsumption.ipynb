# For second file (05/26/17)

```python
import pandas as pd
import numpy as np

dtype = [np.float64, np.int32, 'str'] + 96 * [np.float64] #defining the data types for the columns
df = pd.read_csv('/Users/nandinimalik/Downloads/lenfest_summer_data/Power/LenfestPowerData20170526.csv')
col_names =df.columns
dtype = dict(zip(col_names, dtype))
df2 = pd.read_csv('/Users/nandinimalik/Downloads/lenfest_summer_data/Power/LenfestPowerData20170526.csv', na_values = [' ']) #setting the spaces as the null values

del df2['SERIALNO']
del df2['MCSLOT']

columns = [x for x in range(98)] #changing names of columns to integer values for convenience 
df2.columns = columns
for k in range(1,96):
    df2[k]=df2[k+1]-df2[k]

df2 = df2.groupby(by = 0).sum() #combining values of phase 1 and phase 2 for each apartment
df2.head()
```

```python
# function 1 to remove null values

rows = []
for apartment,j in df2.iterrows():#iterating through the rows
    rows.append(apartment)
#print rows
for i in rows:
    for column in df2.columns[1:]: #iterating through the columns
        if df2.loc[i,column] == float('nan')  : #if a particular entry is a nan value, do
            a = df2.loc[i,column-1] #setting a = the entry in the previous column in the same row 
            c = column
            while df2.loc[i,c] == float('nan'): #finding the next non-null value in the row
                c += 1
                continue
            df2.loc[i,column] = (a + df2.loc[i,c])/2 #setting the null value equal to the average of the provious and next non null value
        else:
            continue
```

```python
df2 = df2.transpose()
df2.info()
```

```python
df2['Date'] = [pd.to_datetime('05-26-2013')]*97 #adding date column

df2.set_index('Date')
df2['Interval'] = [x for x in range(1,98)] #adding column with interval number

df2 = df2.reindex_axis(['Interval'] + list([a for a in df2.columns if a != 'Interval']), axis=1)
df2 = df2.reindex_axis(['Date'] + list([a for a in df2.columns if a != 'Date']), axis=1)

df2.head()
```
